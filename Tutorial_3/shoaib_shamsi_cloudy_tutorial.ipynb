### Planetary Nebula
First up, we will simulate the SED for a planetary nebula region for different abundances. We will begin by creating the simulation inputs.

First, a simple helper function that will allow us to write a string to a text file in the current directory of this tutorial notebook.

```python
def write2file(text, filename):
    file_obj = open(filename, 'w')
    file_obj.write(text)
    file_obj.close()
```

Perfect. We'll write the simulation inputs for our PNe to a file called `pn.in`.

```python
pn_input_pn_abnd = """blackbody, T=1e5 K    # The continuum shape will be a blackbody of temperature 10^5 K
luminosity total 38    # The luminosity of the star in log erg s-1
radius 18    # The log of the inner radius in cm
hden 5    # The log of Hydrogen density in cm-3
sphere    # The geometry of the cloud is a sphere surrounding the emitter
abundances planetary nebula    # The chemical abundances of a typical planetary nebula are set
save overview ".ovr"    # Create an overview file for the simulation
save continuum ".con" units microns    # Create a continnum information file for the simulation"""
```

Now let's gather the wavelength vs. flux data we've just simulated.

```python
pn_abundances = ['pn', 'solar', 'prim', 'crab', 'hii', 'ism']

wav_arrays = []
I_arrays = []
for abundance in pn_abundances:
    wav_arr = (pd.read_table('pn_{}_abnd.con'.format(abundance)))['#Cont  nu'].to_numpy()
    I_arr = (pd.read_table('pn_{}_abnd.con'.format(abundance)))['total'].to_numpy()

    wav_arrays.append(wav_arr)
    I_arrays.append(I_arr)
```

And plotting it reveals:

```python
pn_abundance_labels = ['PNe Abundances', 'Solar Abundances', 'Primordial Abundances', 'Crab Nebula Abundances',
                       'HII Region Abundances', 'Warm-Cold Averaged ISM Abundances']

plt.figure(figsize=(11, 7))

for wav_arr, I_arr, label in zip(wav_arrays, I_arrays, pn_abundance_labels):
    plt.plot(wav_arr, I_arr, label=label, linewidth=0.7)

plt.yscale('log')
plt.xscale('log')

plt.xlim(1e-1, 1e3)
plt.ylim(1e33, 1e41)

plt.xlabel("Wavelength [$\mu$m]")
plt.ylabel("Flux $\\nu$ $f_{\\nu}$ [erg cm$^{-2}$ s$^{-1}$]")

plt.title("SED from Simulations of PNes with Various Abundances", size=20)

plt.legend()
plt.show()
```

This result is super interesting. We can see that using HII region abundances, planetary nebula abundances, and averaged ISM abundances present us with the IR emission which we expect from dust grains. Indeed, both the planetary nebula and averages ISM abundances use the same dust grain settings in the Cloudy code. Inspecting the HII abundance file reveals that it uses grain settings from the Orion Nebula.

It is interesting to see that the Crab nebula abundances do not include dust grain settings, which results in no IR blackbody emission. Although it is known that the Crab nebula, like similar SNRs, contains dust ejecta.

### Broad Line Regions

We will now run a simulation a BLR. Let's write the input file.

```python
blr_input = """table power law    # Using a build in power table law to simulate quarar emission
phi(H) 18.5    # The log of the flux of H-ionizing photons in [cm-2 s-1]
hden 10    # log of hydrogen density in cm-3
stop column density 22    # log of hydrogen column density in cm-2
iterate to convergence    # Iterate until optical depths converge
save overview ".ovr" last    # Output only the last iteration overview
save continuum ".con" units microns last    # Output only the last iteration continuum"""
```

### Plotting Distance vs. Gas Temperature for PNes

We'll use the `.ovr` files to obtain radius and temperature.

```python
pd.read_table('pn_pn_abnd.ovr')
```

We will gather the radius and temperature arrays from our simulations

```python
radius_arrays = []
temp_arrays = []

for abundance in pn_abundances:
    radius_arr = (pd.read_table('pn_{}_abnd.ovr'.format(abundance)))['#depth'].to_numpy()
    temp_arr = (pd.read_table('pn_{}_abnd.ovr'.format(abundance)))['Te'].to_numpy()

    radius_arrays.append(radius_arr)
    temp_arrays.append(temp_arr)
```

Let us now make our plot

```python
plt.figure(figsize=(11, 7))

for radius_arr, temp_arr, label in zip(radius_arrays, temp_arrays, pn_abundance_labels):
    plt.plot(radius_arr, temp_arr, label=label, linewidth=0.7)

plt.xscale('log')

plt.xlabel("Radius [cm]")
plt.ylabel("Temperature [K]")

plt.title("Gas Temperature vs. Radius from our Simulations of PNes with Various Abundances", size=14)

plt.legend()
plt.show()
```

This in very interesting to see. We know that the simulation is supposed to stop when the optical depth converges. For all lines but one, this occurs when the temperature is around 4000 K. For a PNe with primordial abundances though, the temperature goes to a minimum of 2000 K. Primordial abundances have very little helium comparitively, and the only other elements present besides H and He are sparce amounts of Li and Be. Parts of this light gas must have had the time to expand and reach this low temperature before optical depths converged for it.
